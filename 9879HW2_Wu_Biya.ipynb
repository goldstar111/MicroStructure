### 2. (4 points) 
Explain why the optimal market/limit order decision depends in general on quantity available on both sides of the order book in the Parlour model.
#### Answer
-   <span>An increase in book depth $n_B$ on the bid-side decreases the
    probability of a limit buy order.</span>
    - Given that new traders pending limit order have minor chance of execution

-   <span>A decrease in book depth $n_A$ on the ask-side decreases the
    probability of a market sell order in subsequent periods and so
    decreases the probability of a limit buy order.</span>
    - The shallow queue makes limit order easier to execute, in turn decreases the motivation of market order

-   <span>Order signs will be autocorrelated - a <span>*herding*</span>
    effect.</span>
    - A market sell order increases the probability of a subsequent market sell and decreases the probability of a
        subsequent market buy.
    - For instance, a market sell order can devour the corresponding limit buy order, decreasing the buy queue depth,         which makes the sell queue relatively longer. The new seller driven by the ostensible longer sell queue may           be more adamant to execute market sell order.

When $\lambda(u)=\alpha$,  a constant, the BMP order density becomes

\begin{eqnarray*}
\rho_{BMP}(\Delta) &=& \frac{\alpha}{\delta}\,\left\{e^{-\Delta/{\tilde \Delta}}\,\int_0^{\Delta/{\tilde\Delta}}\,\sinh(u)\,du + \sinh(\Delta/{\tilde \Delta})\,\int_{\Delta/{\tilde\Delta}}^\infty\,e^{-u}\,du\right\} \\
&=&\frac{\alpha}{\delta}\,\left\{1 - e^{-\Delta/{\tilde \Delta}}  \right\}
\end{eqnarray*}

where the characteristic distance

$$
\tilde \Delta=\sqrt{\frac{\sigma^2}{2\,\delta}}.
$$
