### 1. (18 points)
Following Almgren and Chriss,  assume that the stock price $S_t$ evolves as

$$
dS_t = \sigma\,dZ_t
$$

and the price $\tilde S_t$ at which we transact is given by

$$
\tilde S_t = S_t - \eta\,v_t
$$

where $v_t:=-{\dot x}_t$ is the rate of trading with $x_0 = X$ and $x_T = 0$.

In the lecture notes, we showed that with a risk term that penalizes average VaR instead of variance, the risk-adjusted cost of trading associated with a given price path $\{S_t\}$ is given by

$$
C = \int_0^T\,(S_t - \eta\,v_t) \,dx_t + \lambda\,\sigma\,\int_0^T\,x_t\,dt
$$

for some price of risk $\lambda$. We want to find a (possibly state-dependent) control $v_t$ that minimizes the expected cost $\cC=\E[C]$.

(a) Apply the technique of integration by parts to simplify the term $\int_0^T S_t dx_t$. Write down the HJB equation for the resulting problem.

**Answer**

$$\int_0^T S_t dx_t = \left. S_t x_t \right|_0^T -\int_0^T x_t dS_t = -S_0 X - \sigma \int_0^T x_t dZ_t$$
\begin{eqnarray*}
\E[C] &=& \E[\int_0^T\,(S_t - \eta\,v_t) \,dx_t + \lambda\,\sigma\,\int_0^T\,x_t\,dt ]\\
&=&\E[\int_0^T S_t dx_t - \int_0^T \eta\,v_t dx_t+ \lambda\,\sigma\,\int_0^T\,x_t\,dt ]\\
&=&\E[-S_0 X - \sigma \int_0^T x_t dZ_t+ \eta\,\int_0^T v_t^2 dt+ \lambda\,\sigma\,\int_0^T\,x_t\,dt]\\
&=&-S_0 X +\E[\int_0^T （\eta \dot x_t^2+ \lambda\,\sigma x_t） dt]
\end{eqnarray*}
substitue into
$$
\p_t J(t,x) + \min_{v\in G[t]} \{\mathcal L^{(v)} J(t,x) + h(t,x,v) \} = 0, \mbox{ for } t < T
$$
so we get 
$$
C_t+\lambda \sigma x + \min\{- C_x v + \eta v^2\}=0
$$


(b) Solve the first order condition to find the optimal trading rate $v^\star$.    

**Answer**

$$v^\star = \frac{C_x}{2 \eta}$$



(c) Substitute $v^\star$ back into the HJB equation to show that the value function $\cC$ must satisfy

$$
\cC_t = \frac{\cC_x^2}{4\,\eta\,} - \lambda \sigma x.
$$

**Answer**
$$
C_t+\lambda \sigma x + (- C_x \frac{C_x}{2 \eta} + \eta (\frac{C_x}{2 \eta})^2)=0
$$
so we get 
$$
\cC_t = \frac{\cC_x^2}{4\,\eta\,} - \lambda \sigma x.
$$

(d) With the terminal condition, 

$$\lim_{t\uparrow T}\cC(t,x) = \begin{cases}0&\text{if $x=0$,}\\
+\infty&\text{if $x\neq0$.}
\end{cases}$$

solve the HJB equation with the ansatz $\cC(t,x) = \eta\,\left[a(t)\,x^2 + b(t)\,x + c(t)\right]$. Verify that the optimal trading rate is given by

$$
v^\star = \frac{x}{T-t} + \frac{1}{4}\, \kappa \, (T-t)
$$

with 

$$
\kappa = \frac{\lambda\,\sigma}{\eta}.
$$

**Answer**

$$
C_t = \eta\,\left[\dot a(t) x^2 +\dot b(t)\,x +\dot c(t)\right] \quad C_x=\eta\,\left[2 a(t)\,x + b(t)\right]
$$

Compare the coefficients and obtain the following system of ODEs

\begin{eqnarray*}
\dot a &=& a^2, \quad a(T)=\infty\\
\dot b &=& ab - \frac{\lambda \sigma}{\eta}, \quad b(T)=\infty\\
\dot c &=& \frac{b^2}{4}, \quad c(T)=0
\end{eqnarray*}

After solving, we get $a = \frac{1}{T-t}\quad b = \frac{\lambda \sigma}{2\eta}(T-t) \quad c=\frac{\lambda^2 \sigma^2 (T-t)^3}{48 \eta^2}$

And 
\begin{eqnarray*}
v^\star &=& \frac{C_x}{2 \eta} =  \frac{\eta\,\left[2 a(t)\,x + b(t)\right]}{2 \eta}=a(t)\,x + \frac{b(t)}{2}\\
&=& \frac{x}{T-t} + \frac{\kappa }{4} (T-t) \quad \text{with} \quad \kappa = \frac{\lambda\,\sigma}{\eta}
\end{eqnarray*}

(e) Define the characteristic time $T^\star$ as

$$
T^\star=\sqrt{\frac{4\,X}{\kappa}}=\sqrt{\frac{4\,\eta\,X}{\lambda\,\sigma}}.
$$


Verify that the optimal trading strategy with liquidation horizon $T$ set to the characteristic time $T^\star$ is given by

$$
\begin{eqnarray*}
x_t&=&X\,\left(1-\frac{t}{T}\right)^2.
\end{eqnarray*}
$$

**Answer**

\begin{eqnarray*}
-\frac{dx}{dt}&=&\frac{x}{T-t} + \frac{\kappa }{4} (T-t) \quad \text{with} \quad \kappa = \frac{\lambda\,\sigma}{\eta}\\
&=>& -(T-t)dx=x dt+ \frac{\kappa }{4} (T-t)^2 dt\\
&=>& (t-T)dx-xd(t-T)=\frac{\kappa }{4} (T-t)^2 dt\\
&=>& \frac{(t-T)dx-xd(t-T)}{(t-T)^2}=\frac{\kappa }{4} dt\\
&=>& d(\frac{x}{t-T})=\frac{\kappa }{4} dt\\
&=>& \frac{x}{t-T} =\frac{\kappa }{4} t + C_1 \\
&=>& x(t) =\frac{\kappa }{4}(t-T)^2+C_2(t-T)  \quad \text{with}\, x(0)=X \,and\, x(T)=0\\
&=>& x(t) =\frac{\kappa }{4}(t-T)^2 \quad\text{while} \,C_2=-\frac{X}{T}+\frac{\kappa T}{4}=-\frac{X}{T}+\frac{\frac{\lambda\,\sigma}{\eta} T}{4}=0\\
&=>& x(t) =\frac{\kappa T^2}{4}(1-\frac{t}{T})^2 = \frac{ \frac{\lambda\,\sigma}{\eta} T^2}{4}(1-\frac{t}{T})^2=X\,\left(1-\frac{t}{T}\right)^2
\end{eqnarray*}

(a) The *msft.bats* dataset from *MSFT130311.rData* contains all trades on the BATS exchange on 11-Mar-2013.  One trading days has 390 minutes.  Sample the  *msft.bats* dataset evenly, roughly every minute of volume time.  Superimpose plots of the optimal strategy under ABM and GBM assuming liquidation takes place over one trading day.


```r
df<-msft130311.bats
cumvol <- cumsum(abs(df$signed.shares))
interval<- tail(cumvol,1)/390
px_last<-numeric(390)
for (i in 1:390)
{
    px_last[i]<-df$trade.price[findInterval(interval*i,cumvol)]
}

T<-390
dt<-1
X<-100

# ABM
xt<-X
va <- numeric(T-1)
for (t in 1:(T-1))
{
    va[t]<-xt/(T-t)+X/T*(1-t/T)
    xt<-xt-va[t]*dt
}

#GBM
xt<-X
vg <- numeric(T-1)
for (t in 1:(T-1))
{
    vg[t]<-xt/(T-t)+X/T*px_last[t]/head(df$trade.price,1)*(1-t/T)
    xt<-xt-vg[t]*dt
}

plot(1:389, va, xlab = "Time Stamps", ylab = "Trading Rates", col = "red", type = "p", main = 'ABM & GBM')
lines(1:389, vg, col = "green", type = "l", lwd = 3)

legend("topright", c("ABM","GBM"), col = c("red","green"), lwd = 1)
```


(b) Denote the corresponding position sizes by $x^A(t)$ and $x^G(t)$ respectively.  What is the maximum deviation of the two position sizes as a percentage of the initial position $X$?  How significant is this deviation?



```r
xta<-X
xtg<-X
va <- numeric(T-1)
vg <- numeric(T-1)
diffxt<-numeric(T-1)
for (t in 1:(T-1))
{
    va[t]<-xta/(T-t)+X/T*(1-t/T)
    xta<-xta-va[t]*dt
    vg[t]<-xtg/(T-t)+X/T*px_last[t]/head(df$trade.price,1)*(1-t/T)
    xtg<-xtg-vg[t]*dt
    diffxt[t]<-xta - xtg
}
cat("the maximum disparity is ",max(abs(diffxt),"\n"))
t.test(diffxt)
cat("the deviation is significantly below zero")
```

## Kratz and Schöneborn with time-averaged VaR risk charge

### 3. (12 points)

(a) Repeat the computation of the optimal strategy from the lecture notes with time-averaged VaR

$$
\lambda\,\sigma\,\int_0^T\,x_t\,dt
$$

as the risk term (rather than the quadratic variation $\lambda\,\sigma^2\,\int_0^T\,x_t^2\,dt$).  Show that the optimal position

$$
x^\star(t)=\left(X-\frac{A\, T}{\theta }\right)\,\frac{1-e^{-\theta\,(T-t)} }{1-e^{-\theta\,T }}+\frac{A }{\theta }\,(T-t)
$$
and deduce the optimal trading rate $v^\star(t)$.



#### Answer

$$\frac{\partial \mathcal{C} }{\partial t}+\lambda \,\sigma\, x+\min_{v,y \in \mathcal{G} }\left\{ -v \,\mathcal{C} _x+\eta\,v^2 +\theta\,\left[\mathcal{C} (x_t-y)-\mathcal{C} (x_t)\right]\right\} =0$$
The value of $y$ that minimizes
$\left[\mathcal{C} (x_t-y)-\mathcal{C} (x_t)\right]$ is obviously
$y^\star=x_t$. <br\>
Differentiating wrt $v$, we get the first order condition on $v$ we get $v^\star=\frac{1}{2\,\eta}\,\mathcal{C} _x.$

$$\frac{\partial \mathcal{C} }{\partial t}+\lambda \,\sigma\, x -\frac{1}{4\,\eta}\,(\mathcal{C} _x)^2 -\theta\,\mathcal{C} =0$$
with terminal condition:
$$\lim_{t\uparrow T}\cC(t,x) = \begin{cases}0&\text{if $x=0$,}\\
+\infty&\text{if $x\neq0$.}
\end{cases}$$
solve the HJB equation with the ansatz $C(t,x)=a(t)x^2+b(t)x+c(t)$. 

So we get 
\begin{eqnarray*}
\dot a(t)-\frac{a^2}{\eta}-\theta a&=&0 \quad \text{with}\,a(T)=\infty\\
\dot b(t)+\lambda \sigma - \frac{ab}{\eta}-\theta b&=&0 \quad \text{with}\, b(T)=\infty\\
\dot c(t)-\frac{b^2}{4\eta}-\theta c&=&0 \quad \text{with}\, c(T)=0
\end{eqnarray*}

After solving we get $$a(t) = \frac{\theta \eta}{e^{\theta(T-t)}-1} \quad b(t)=\frac{\lambda \sigma}{\theta} -\frac{\lambda \sigma(T-t)}{e^{\theta(T-t)}-1} \quad c(t)= \frac{\lambda^2\sigma^2\left\{e^{-\theta(T-t)}\left[\theta^2(T-t)^2+1\right]-1\right\}}{4\eta \theta^2 e^{\theta t}\left(e^{-\theta (T-t)}-1\right)}$$
Thus,
$$
\mathcal{C}_x=2a(t)x+b(t)=\frac{2\eta\theta e^{-\theta(T-t)}x}{1-e^{-\theta(T-t)}}+\frac{\lambda \sigma}{\theta} -\frac{\lambda \sigma(T-t)}{e^{\theta(T-t)}-1}
$$
$$
v^\star=-\dot x=\frac{\mathcal{C}_x}{2\eta}=\frac{\theta e^{-\theta(T-t)}x}{1-e^{-\theta(T-t)}}+\frac{\lambda\sigma}{2\eta\theta}-\frac{\lambda\sigma (T-t)}{2\eta\left[e^{\theta(T-t)}-1\right]}
$$
With initial condition $x(0)=X$, we get
$$
x^\star(t)=\left(X-\frac{A\, T}{\theta }\right)\,\frac{1-e^{-\theta\,(T-t)} }{1-e^{-\theta\,T }}+\frac{A }{\theta }\,(T-t)
$$
where $A=\frac{\lambda\sigma}{2\eta}$
$$
v^\star(t)=\frac{\theta e^{-\theta(T-t)}(X-\frac{AT}{\theta})}{1-e^{-\theta T}}+\frac{A}{\theta}
$$



(b) By Taylor-expanding the solution around $\theta=0$, show that we retrieve the solution derived in the slides with no dark pool

$$
x^\star(t)=\left(X-\frac{A \, T}{2}\,t\right)\,\left(1-\frac{t}{T}\right) .
$$

in the limit $\theta \to 0$.


#### Answer
\begin{eqnarray*}
x^\star(t)&=&\left(X-\frac{A\, T}{\theta }\right)\,\frac{1-e^{-\theta\,(T-t)} }{1-e^{-\theta\,T }}+\frac{A }{\theta }\,(T-t)\\
&=&\frac{(X\theta-AT)(1-e^{-\theta\,(T-t)})+A(T-t)(1-e^{-\theta T})}{\theta(1-e^{-\theta T})}\\
&=&_{L'Hospital's rule}\frac{X(1-e^{-\theta(T-t)})-X\theta e^{-\theta(T-t)}+AT^2 e^{-\theta T}}{1-e^{-\theta T}+\theta T e^{-\theta T}} \\
&=&_{L'Hospital's rule}\frac{(2X-ATt)(T-t)}{2T}\\
&=&\left(X-\frac{A \, T}{2}\,t\right)\,\left(1-\frac{t}{T}\right) 
\end{eqnarray*}
